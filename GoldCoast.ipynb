```python
# vector search a bounding box
def search_unordered(bbox, _type, count=100, cloud_cover=50):
    aoi = box(*bbox).wkt
    query = "item_type:{} AND item_type:LandsatAcquisition".format(_type)
    query += " AND attributes.cloudCover_int:<{}".format(cloud_cover)
    return gbdx.vectors.query(aoi, query, count=count)
```

```python
# vector search with WKT
def search_unordered_wkt(wktstr, _type, count=10, cloud_cover=50):
    aoi = wktstr 
    query = "item_type:{} AND item_type:LandsatAcquisition".format(_type)
    query += " AND attributes.cloudCover_int:<{}".format(cloud_cover)
    return gbdx.vectors.query(aoi, query, count=count) 
```

```python
wktstr = "POLYGON((147.1 -38.16, 129.29 -26.51, 142.43 -10.53, 160.24 -23.71, 147.1 -38.16))"
records = search_unordered_wkt(wktstr, 'LandsatAcquisition',200,25)
ids = [r['properties']['attributes']['catalogID'] for r in records]
#print(records)
```
